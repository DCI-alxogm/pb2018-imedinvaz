```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as cons
%matplotlib inline
```

```python
n = 'IRCF1.txt'
c = np.dtype([('frecuencia cm^-1', 'f'), ('espectro MJy/sr', 'f'),('incertidumbre kJy/sr','f')])
a = np.loadtxt(n, dtype=c, usecols=(0,1,3))
a
x=a['frecuencia cm^-1']
y=a['espectro MJy/sr']*(pow(10,-26))
i=(a['incertidumbre kJy/sr']*(pow(10,-26)))/1000
plt.errorbar(x,y,i)
plt.title("Grafica de frecuencia contra espectro")  
plt.xlabel("Frecuencia cm^-1")  
plt.ylabel("Espectro W*m^-2*Hz*sr^-1")
```
