```python
def recta(a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))
plt.plot(x,recta(x,3,-2))
plt.plot(x,recta(x,-(1/3)))
plt.grid()
```

```python
def recta2(a,m=1,b=0,label='x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return 
```

```python
x=np.linspace(-2,2,20)
recta2(x)
recta2(x,-2,-1,'-x+1')
plt.legend()
plt.grid()
```

```python
def recta2(a,m=1,b=0,label='x'):
    if m<0:
        print('pendiente encontrada, se encuentra cero')
        m=0;
    z=m*a+b
    plt.plot(a,z,label=label)
    return 
```

```python
x=np.linspace(-2,2,20)
recta2(x)
recta2(x,-2,-1,'-x+1')
plt.legend()
plt.grid()
```

```python
##solución de la ecuación de Euler
```

```python
V=np.zeros(10000)
O=np.zeros(10000)
h=0.001
W=2
V[0]=0
O[0]=90
```

```python
def va(O,h,w=1):
    return -(O*h*w**2)
def Oa(V,h):
    return V*h

for i in range(9999):
    V[i+1]=V[i]+va(O[i],h)
    O[i+1]=O[i]+Oa(V[i],h)
t=np.zeros(10000)
t[0]=0.001
for s in range(9999):
    t[s+1]=t[s]+0.001
```

```python
plt.plot(V,O,color='violet')
plt.title("Grafica de velocidad contra ángulo")  
plt.xlabel("Velocidad")  
plt.ylabel("Ángulo")  
```

```python
plt.plot(t,V,label='oscilador',color='violet')
plt.legend()
plt.title("Grafica de velocidad contra tiempo")  
plt.xlabel("Tiempo")  
plt.ylabel("Velocidad") 
```

```python
plt.plot(t,O,label='oscilador',color='violet')
plt.legend()
plt.title("Grafica de velocidad contra tiempo")  
plt.xlabel("Tiempo")  
plt.ylabel("Ángulo") 
```
